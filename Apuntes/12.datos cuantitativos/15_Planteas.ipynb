 El .groupby en python es similar al aggregate o by, donde variable(s)~factor(es) solo que de la forma
 
df.groupby('factor(es)')['columna(s)'].FUN
 

```python
planets.groupby('method')["orbital_period"].median()
```

```python
#otros ejemplos
planets.groupby('method')[["orbital_period",'year']].median()
```

```python
#otros ejemplos
planets.groupby(['year','method'])[["orbital_period"]].median().unstack()
```

# creando factores
supongamos que queremos abrupar por decada, para ello primero hay que crear dicha variable, tomando a√±o, lo asignaremos a una decada




```python
decada = planets.year//10 *10
decada
```

```python
planets.groupby([decada,'method'])[["orbital_period"]].sum().unstack()
```

```python
# incluso podriamos hacerlo como coloquialmente se hace con "80s"
decada = decada.astype(str) +'s'
decada
```

```python
planets.groupby([decada,'method'])[["orbital_period"]].sum().unstack()
```

```python
# notemos que la va decada no tiene mas que el nombre heredado year, por tanto podemos cambiarlo con
decada.name = "decada"
planets.groupby([decada,'method'])[["orbital_period"]].sum().unstack()
```

```python
# quitemos los na
planets.groupby([decada,'method'])[["number"]].sum().unstack().fillna(0)
```
