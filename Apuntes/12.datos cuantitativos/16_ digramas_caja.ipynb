```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(19880519)
```

```python
dist1 = 100 * np.random.rand(50)
dist2 = 50 * np.ones(25)
dist3 = 100 + 100*np.random.rand(10)
dist4 = -100 + np.random.rand(10)

data = np.concatenate((dist1,dist2,dist3,dist4))
```

```python
data
```

```python
data.shape
```

```python
#creando boxplot
fig1,ax1 = plt.subplots()
ax1.set_title('boxplot')
ax1.boxplot(data)
```

```python
#con notch
fig2,ax2 = plt.subplots()
ax2.set_title('boxplot')
ax2.boxplot(data, notch= True)
```

```python
#cambiando color outliers
green_diamond = dict(markerfacecolor = "g", marker = "D")
fig3,ax3 = plt.subplots()
ax3.set_title('boxplot')
ax3.boxplot(data, notch= True, flierprops = green_diamond)
```

```python
#sin outliers
green_diamond = dict(markerfacecolor = "g", marker = "D")
fig3,ax3 = plt.subplots()
ax3.set_title('boxplot')
ax3.boxplot(data, notch= True, showfliers = False)
```

```python
# horizontal
#cambiando color outliers
red_square = dict(markerfacecolor = "r", marker = "s")
fig3,ax3 = plt.subplots()
ax3.set_title('boxplot')
ax3.boxplot(data, notch= True, flierprops = red_square, vert = False )
```

```python
# modificando tamao de bigotes, por defecto 1.5 RIQ
red_square = dict(markerfacecolor = "r", marker = "s")
fig3,ax3 = plt.subplots()
ax3.set_title('boxplot')
ax3.boxplot(data, notch= True, flierprops = red_square, vert = False , whis = 1)
```
