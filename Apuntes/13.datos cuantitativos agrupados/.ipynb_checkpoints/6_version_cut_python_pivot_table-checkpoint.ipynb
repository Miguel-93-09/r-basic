hay diferencias entre groupby y tablas de pivotaje

```python
import numpy as np
import seaborn as sb
import pandas as pd
```

```python
titanic = sb.load_dataset("titanic")
titanic.head()
```

```python
# con grupby podemos hacer tablas acorde a un parametro, añadir numeric only ya que luego de pandas 2.0 se modifico.
titanic.groupby(["sex"]).mean(numeric_only=True)
```

```python
# podemos elegir varaibles sobre cuales se haga dicho gropuby
titanic.groupby(["sex"])['survived'].mean()
```

```python
#se puede con un renglon y una columa si elegimos dos parametros y una variable
titanic.groupby(["sex", "pclass"])['survived'].mean()
```

```python
#pero vemos que si se quiere algo mas parecido a una tabla de pivotaje podemos hacer unstack()
titanic.groupby(["sex", "pclass"])['survived'].mean().unstack()
```

```python
# esto ultimo ya empieza a ser un poco coplicado ,para ello esta ..pivot.table, primero se indica la variable, luego en index los renglones 
#y en columns columnas

titanic.pivot_table("survived", index = "sex", columns = "pclass")
```

# cut

```python
# es parecido a R pd.cut($columna, bins = [] , include_lowest = $T,F , labels = [] , precisioon = , right =  ) presionar tab para más 
age = pd.cut(titanic['age'], bins = [0,18,25,35, 50, 65, 100], include_lowest = True , precision = 1)
age
#veamos que al hacer include lowest, el primer intervalo se baja un poco, precision, señala numero de decimales a considerar
```

```python
# otro ejemplo con intervalos cerrados izquierda, el lowest pierde sentido por que ya se incluye el menor dato
age = pd.cut(titanic['age'], bins = [0,18,25,35, 50, 65, 100], include_lowest = True , precision = 1, right = False)
age
```

```python
# se puede usar esta variable incluso en pivot_table
titanic.pivot_table('survived', index = ['sex',age], columns = 'class')
```

```python
#tambien se puede hacer el corte por cuantiles
fare = pd.qcut(titanic["fare"], q = 4 , labels = ['Q1','Q2','Q3','Q4'] )
fare
```

```python
#y ocuparlo tambien para una pivot
titanic.pivot_table('survived', index = ['sex',age], columns = [fare,'class'])
```

# generalidades de pivot_ table

df.pivot_table(data, vaules= None, index = None, columns = None, 
                aggfunc = "mean" , fill_value=None , margins = False,
                dropna = True, margins_name= "All")

margins hce los totales, aggfunc, se le puede pasar un diccionario conde la clave es la variable factor y la otra parte del : es la funcion especifica que se le quiere aplicar

```python
# incluso se podria no indicar la variable si esto se indica en aggfunc

titanic.pivot_table(index = ['sex',age], columns = [fare,'class'], aggfunc ={'survived':sum})
```

```python
# o aplicar varias funciones a cada variable 
titanic.pivot_table(index = ['sex'], columns = ['class'], aggfunc ={"fare": "mean" , 'survived': "sum"})
```

```python
# o aplicar varias funciones a cada variable 
titanic.pivot_table("survived" , index = ['sex'], columns = ['class']  , margins = True, margins_name = "total")
```

```python
titanic.pivot_table('survived' ,index = ['sex'], columns = ['class'], margins = True)
```
