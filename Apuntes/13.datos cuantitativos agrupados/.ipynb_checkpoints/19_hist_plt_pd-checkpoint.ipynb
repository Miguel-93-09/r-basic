```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(2019)
x = np.random.laplace(loc = 10, scale = 3, size = 1000)
```

```python
n , bins, patches = plt.hist(x=x, bins = "auto", alpha = 0.75, rwidth = 0.85)
plt.grid(axis = "y" , alpha = 0.5)
plt.xlabel("valor")
plt.ylabel("frecuencia")
plt.text(-10,90,r'$\mu = 10, b=3$')
#extendemos un poco limite de y, fijarse en forma  $resutado1 if $condicion1 else $condicion2
maxfreq = n.max()
plt.ylim( ymax = np.ceil(maxfreq/10) * 10   if maxfreq%10    else    maxfreq + 10  )
```

# histograma con pandas

Permite hacer histogramas pero requiere que se alimente de series y utiliza pyplot

```python
# usamos dist gamma
import pandas as pd
```

```python
size , scale = 1000,10
data = pd.Series(np.random.gamma(scale, size = size))
```

```python
data.plot.hist(grid = True, bins = 20, rwidth = 0.9, color = "#d52675")
plt.grid(axis = "y", alpha = 0.75)
```
