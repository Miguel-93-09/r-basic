```python
#como en la siguiente no se indicia day, month entonces se eliminaran estas varaibales, aplicandose la funcion suma sobre ellas apra "eliminarlas"
births.pivot_table('births', index = 'year', columns='gender', aggfunc= "sum" )
```

```python
# creando columna decada
births['decade'] = births['year'] // 10 * 10

births
```

```python
# pivotaje por decada
births.pivot_table('births',index = 'decade', columns = 'gender' , aggfunc='sum')
```

```python
#vease que columns hace que salgan dos lineas, index sera eje x
births_year = births.pivot_table('births', index = 'year', columns = 'gender', aggfunc = 'sum')
births_year.plot()
```

```python
#mejoramos grafico
plt.figure(figsize =(15,8))
births_year = births.pivot_table('births', index = 'year', columns = 'gender', aggfunc = 'sum')
births_year.plot()
plt.show()
```

```python
#limpiaremos datos
quartiles = np.percentile(births['births'],[25,50,75])
quartiles
```

```python
med = quartiles[1]
IQR = quartiles[2] - quartiles[0]
sig = 0.75 * IQR
```

```python
# filtrando, el @ es para acceder a variables globales entre texto 
births = births.query("(births > @med - 5 * @sig) & (births < @med + 5 * @sig)")
births.shape
```

```python
births.shape
```

```python
# convirtiendo dias a entero, para que salga el formateo de fecha
births['day'] = births['day'].astype(int)
births.head()
```

```python
#cambiando el index por fechas, notar uso de pd
births.index = pd.to_datetime(10000 * births.year + 100 * births.month + births.day , format = "%Y%m%d")
births.head()
```

```python
#agregando columna para agrupar por dia de la semana, probar luego con una pivot table
births['dayofweek'] = births.index.dayofweek
```

```python
births
```

```python
#sacamos promedio con la ultima table
births_day = births.pivot_table('births', index = 'dayofweek' , columns = 'decade' , aggfunc = "mean")
births_day
```

```python
births_day.plot()
```

```python
#alternativa, en este caso el problema es que se toma el dia del mes, en la anterior 0 es lines, 1 martes...
births.pivot_table('births', index = 'day' , columns ='decade' , aggfunc = 'mean')
```

```python
# mejorandola
births_day.plot()
#Cambiemos etiquetas eje x, notemos que dejamos un espacio en blanco o no funcionara
plt.gca().set_xticklabels(['','L','M','X','J','V','S','D'])
plt.ylabel('Media de nacimientos por día')
plt.show()
```

```python
# haremos que se cree una tabla por mes , luego dia
births_by_date = births.pivot_table('births', index = [births.index.month , births.index.day] , aggfunc = 'mean')
births_by_date
```

```python
#crearemos una tabla con indice fecha y births desde la ultima table, notemos que el indice de la ultima tabla es una dupla, esto difiere
#del video porque ya no existe  datetime solo, notemos que no son del 2020 si no de ttodos los años ya que proviene de la ultima table
births_by_date.index = [pd.to_datetime ("2020" + str(month)  + str(day) , format = '%Y%m%d' ) for (month,day) in births_by_date.index ]
```
