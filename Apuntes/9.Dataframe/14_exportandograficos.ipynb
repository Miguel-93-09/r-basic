```python
import numpy as np
import pandas as pd
```

```python
# crenado lista de a√±os
list(range(1900,2020))
```

```python
# otra opcion mas rebuscada
year = [i for i in range(1900,2020)]
deads = [(d + np.random.uniform(0,100)-1850) for d in year]
```

```python
# creando su df
df = pd.DataFrame({
    "year" : year,
    "deads" : deads
})
```

```python
# graficando directamente  con plt
df.plot(x = "year", y = "deads", figsize = (10,8) )
```

```python
# exporando figura
my_plot = df.plot(x = "year", y = "deads", figsize = (10,8))
my_fig  = my_plot.get_figure()
my_fig.savefig("prueba.png")
```
