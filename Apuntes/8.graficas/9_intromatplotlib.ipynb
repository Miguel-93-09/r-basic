```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4]
```

```python
# como en R toma unico vector como si fuera y, pero empieza en 0
plt.plot(x)
```

```python
plt.plot(x)
plt.xlabel("abcisas")
plt.ylabel("ordenadas")
plt.show() #para agregar configuraciones adicionales
```

```python
# dos conjuntos coordenadas como en R
y = [1,4,5,7]
plt.plot(x,y)
plt.show()
```

```python
help(plt.plot)
```

```python
# configurando linea, primero color, luego forma de representacion datos
plt.plot(x,y,"r-")
plt.show()
```

```python
# cambiando encuadre que se muestra  [x_0,x_1,y_0,y_1]
plt.plot(x,y,"ro")
plt.axis([-1,5,-1,10])
plt.show()
```

```python
import numpy as np
data = np.arange(0,10,0.2)
data
```

```python
#graficando varias lineas
plt.plot(data, data, "r--", data , data**2 , "bs", data , data**3 , "g^")
plt.show()
```

```python
# modificando linea
plt.plot(data, data, "r--", linewidth =4.0)
plt.show()
```

```python
# modificando otra spropieades de linea, es importante la coma para sealar que solo extraemos ese atributo
line, = plt.plot(data, data, "g-", linewidth =4.0)
line.set_antialiased(False)
plt.show()
```

```python
# modificando varios con setp
line = plt.plot(data, data, "r--", data , data**2 , "bs", data , data**3 , "g^")
plt.setp(line,color = "r", linewidth=2)
plt.show()
```

```python
# alpha pra transparencia
plt.plot(x,y,"g-",alpha= 0.3)
plt.show()
```

```python
# marker, entonces primero va color en, luego linestyle y por ultimo marker
plt.plot(x,y,"r:+")
```

```python
# esto es equivalente a lo anterior
plt.plot(x,y,color = "b", linestyle = ":" , marker = "+")
```

```python
# si no se pone ningun parametro, solo se pasa el line del caso anterior, nos salen todas las opciones a configurar
plt.setp(line)
```
