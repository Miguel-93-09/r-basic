```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# para crear grafico de funcion
def f(x):
    return np.exp(-x)*np.cos(2*np.pi*x)
```

```python
x1 = np.arange(0,5.1,0.1)
x2 = np.arange(0,5.1,0.2)
```

```python
# se crea la figura, como objeto que contendra los graficos, opcional si solo se crea una figura, por edefecto subplot es (111)
plt.figure(1)

#numerito signfica filas (2), columnas (1) , primer grafico (1),
plt.subplot(211)
plt.plot(x1,f(x1),"ro",x2,f(x2),"k-")

plt.subplot(212)
plt.plot(x2,f(x2), "r--v")

plt.show()
```

```python
#dos figuras

plt.figure(1)

plt.subplot(121)
plt.plot(x1,f(x1),"ro",x2,f(x2),"k")

plt.subplot(122)
plt.plot(x2,f(x2), "rv--")


#queremos modificar algo, aadir, entonces debe ser inmediatamente despues de la figure correspondiente
plt.figure(1)
plt.subplot(121)
plt.title("primer subplot")
plt.subplot(122)
plt.title("segundo subplot")

plt.figure(2)

plt.subplot(211)
plt.plot(x1,f(x1),"ro",x2,f(x2),"k")

plt.subplot(212)
plt.plot(x2,f(x2), "rv--")

plt.show()
```
