```python
%matplotlib inline 
import matplotlib.pyplot as plt
import numpy as np
```

```python
# cambios en ejes de dibujo
x = np.linspace(-np.pi, np.pi, 256, endpoint = True)
```

```python
S ,C =np.sin(x),np.cos(x)
```

```python
#Graficos en mismo eje, tambien se pueden hacer en una misma linea como casos anteriores
plt.plot(x, S, "b-", linewidth = "1.2")
plt.plot(x, C, "g-", linewidth = "1.2")
plt.xlim(-4,4)
plt.ylim(-1.5,1.5)

# modificar donde estan las marquitas de los ejes
plt.xticks(np.linspace(-4,4,10, endpoint = True))
plt.yticks(np.linspace(-1,2,10, endpoint = True))

plt.show()
```

```python
# para expandir ejes mas profesional

plt.plot(x, S, "b-", linewidth = "1.2")
plt.plot(x, C, "g-", linewidth = "1.2")
plt.xlim(x.min()*1.1,x.max()*1.1)
plt.ylim(S.min()*1.1,S.max()*1.1)

# Se pueden modificar ticks a gusto con un vector
plt.xticks([-np.pi,-np.pi/2,0,np.pi/2,np.pi])
plt.yticks(np.linspace(-1,1,10, endpoint = True))

plt.show()
```

```python
# aun mejores ticks en los ejes
plt.plot(x, S, "b-", linewidth = "1.2")
plt.plot(x, C, "g-", linewidth = "1.2")
plt.xlim(x.min()*1.1,x.max()*1.1)
plt.ylim(S.min()*1.1,S.max()*1.1)


plt.xticks([-np.pi,-np.pi/2,0,np.pi/2,np.pi],
            [r'$-\pi$',r'$-\pi/2$',r'$0$',r'$\pi/2$','$2\pi$'])
            
plt.yticks(np.linspace(-1,1,3, endpoint = True),
          ["-1","0","+1"])

plt.show()
```

```python
# modificando ejes y añdiendo labels en primera orden de plot
plt.plot(x, S, "b-", linewidth = "1.2", label = "seno")
plt.plot(x, C, "g-", linewidth = "1.2",  label = "coseno")
plt.xlim(x.min()*1.1,x.max()*1.1)
plt.ylim(S.min()*1.1,S.max()*1.1)


plt.xticks([-np.pi,-np.pi/2,0,np.pi/2,np.pi],
            [r'$-\pi$',r'$-\pi/2$',r'$0$',r'$\pi/2$','$2\pi$'])
            
plt.yticks(np.linspace(-1,1,3, endpoint = True),
          ["-1","","+1"])

#con plt.gca accedemos al eje del plot actual
ax = plt.gca()

#desapareciendo ejes
ax.spines["right"].set_color('none')
ax.spines["top"].set_color('none')

#cambiando posicion de los sitck, eje x, por defecto esta en bottom en este caso
ax.xaxis.set_ticks_position("bottom")

#cambiando posicion ejes x en 0 y  y en 0 tambien
ax.spines['bottom'].set_position(('data',0))
ax.spines['left'].set_position(('data',0))

# se añade leyeda con las etiquetas
plt.legend(loc="upper left")



plt.show()
```

```python
# añadiendo una anotacion por ejeplo

plt.plot(x, S, "b-", linewidth = "1.2", label = "seno")
plt.plot(x, C, "g-", linewidth = "1.2",  label = "coseno")
plt.xlim(x.min()*1.1,x.max()*1.1)
plt.ylim(S.min()*1.1,S.max()*1.1)


plt.xticks([-np.pi,-np.pi/2,0,np.pi/2,np.pi],
            [r'$-\pi$',r'$-\pi/2$',r'$0$',r'$\pi/2$','$2\pi$'])
            
plt.yticks(np.linspace(-1,1,3, endpoint = True),
          ["-1","","+1"])


ax = plt.gca()


ax.spines["right"].set_color('none')
ax.spines["top"].set_color('none')


ax.xaxis.set_ticks_position("bottom")


ax.spines['bottom'].set_position(('data',0))
ax.spines['left'].set_position(('data',0))


plt.legend(loc="upper left")

#empeiza aqui ejemplo
x0=2* np.pi /3
plt.plot([x0,x0],[0,np.sin(x0)],"b--", linewidth=2.5)
plt.scatter([x0,],[np.sin(x0),],50,"b")

#añadiendo anotacion, en xycoords se cambia la medicion relativa a desde las coordenadas y no el punto, textcoords no lo entiendo
#invectigar depsues, por ahora funciona bien sin el mismo
#connection style curva con rad la flecha, arc3 no se que sea

plt.annotate(r'$\sin(\frac{2\pi}{3})$',
             xy=(x0,np.sin(x0)),
             xytext=(x0+0.5,np.sin(x0)+0.1),
             xycoords ="data",
             fontsize = 16,
              arrowprops = dict(arrowstyle = "->",connectionstyle = "arc3,rad=.5")
             )



plt.show()
```

```python
#ejemplo otra flecha 
# añadiendo una anotacion por ejeplo

plt.figure(figsize=(10,8))

plt.plot(x, S, "b-", linewidth = "1.2", label = "seno")
plt.plot(x, C, "g-", linewidth = "1.2",  label = "coseno")
plt.xlim(x.min()*1.1,x.max()*1.1)
plt.ylim(S.min()*1.1,S.max()*1.1)


plt.xticks([-np.pi,-np.pi/2,0,np.pi/2,np.pi],
            [r'$-\pi$',r'$-\pi/2$',r'$0$',r'$\pi/2$','$2\pi$'])
            
plt.yticks(np.linspace(-1,1,3, endpoint = True),
          ["-1","","+1"])


ax = plt.gca()


ax.spines["right"].set_color('none')
ax.spines["top"].set_color('none')


ax.xaxis.set_ticks_position("bottom")


ax.spines['bottom'].set_position(('data',0))
ax.spines['left'].set_position(('data',0))


plt.legend(loc="upper left")

#empeiza aqui ejemplo
x0=2* np.pi /3
plt.plot([x0,x0],[0,np.sin(x0)],"b--", linewidth=2.5)
plt.scatter([x0,],[np.sin(x0),],50,"b")


plt.annotate(r'$\sin(\frac{2\pi}{3})$',
             xy=(x0,np.sin(x0)),
             xytext=(x0+0.5,np.sin(x0)+0.1),
             xycoords ="data",
             fontsize = 16,
              arrowprops = dict(arrowstyle = "->",connectionstyle = "arc3,rad=.5")
             )

plt.plot([x0,x0],[0,np.cos(x0)],"g--", linewidth=2.5)
plt.scatter([x0,],[np.cos(x0),],50,"g")


#segundo punto

plt.annotate(r'$\cos(\frac{2\pi}{3})$',
             xy=(x0,np.cos(x0)),
             xytext=(x0+0.5,np.cos(x0)+0.1),
             xycoords ="data",
             fontsize = 16,
              arrowprops = dict(arrowstyle = "->",connectionstyle = "arc3,rad=.5"))



plt.show()

```

```python
#corrigiendo etiquetas para que sea vean bien
#ejemplo otra flecha 
# añadiendo una anotacion por ejeplo

plt.figure(figsize=(10,8))

plt.plot(x, S, "b-", linewidth = "1.2", label = "seno")
plt.plot(x, C, "g-", linewidth = "1.2",  label = "coseno")
plt.xlim(x.min()*1.1,x.max()*1.1)
plt.ylim(S.min()*1.1,S.max()*1.1)


plt.xticks([-np.pi,-np.pi/2,0,np.pi/2,np.pi],
            [r'$-\pi$',r'$-\pi/2$',r'$0$',r'$\pi/2$','$2\pi$'])
            
plt.yticks(np.linspace(-1,1,3, endpoint = True),
          ["-1","","+1"])


ax = plt.gca()


ax.spines["right"].set_color('none')
ax.spines["top"].set_color('none')


ax.xaxis.set_ticks_position("bottom")


ax.spines['bottom'].set_position(('data',0))
ax.spines['left'].set_position(('data',0))


plt.legend(loc="upper left")

#empeiza aqui ejemplo
x0=2* np.pi /3
plt.plot([x0,x0],[0,np.sin(x0)],"b--", linewidth=2.5)
plt.scatter([x0,],[np.sin(x0),],50,"b")



plt.annotate(r'$\sin(\frac{2\pi}{3})$',
             xy=(x0,np.sin(x0)),
             xytext=(x0+0.5,np.sin(x0)+0.1),
             xycoords ="data",
             fontsize = 16,
              arrowprops = dict(arrowstyle = "->",connectionstyle = "arc3,rad=.5")
             )

plt.plot([x0,x0],[0,np.cos(x0)],"g--", linewidth=2.5)
plt.scatter([x0,],[np.cos(x0),],50,"g")


plt.annotate(r'$\cos(\frac{2\pi}{3})$',
             xy=(x0,np.cos(x0)),
             xytext=(x0+0.5,np.cos(x0)+0.1),
             xycoords ="data",
             fontsize = 16,
              arrowprops = dict(arrowstyle = "->",connectionstyle = "arc3,rad=.5"))


#bbox colorea la cajita al rededor del numero de las sitcks

for label in ax.get_xticklabels() + ax.get_yticklabels():
    label.set_fontsize(16)
    label.set_bbox(dict(facecolor="white",edgecolor="None",alpha = 0.6))

plt.show()
```
