```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
mu = 100
sigma = 20

x = mu + sigma * np.random.randn(10000)
```

```python
n, bins, patches = plt.hist(x, 50, density=1 , facecolor = "g", alpha = 0.8)
plt.xlabel(r"peso $N$")
plt.ylabel("proba")

#modificando detalles titulos, tectos
plt.title("histo 1", fontsize = 14, color = "g")

#añadiendo texto en grafico, para latex usar : r"text $latex$ "
plt.text(40,0.015, r'$\mu = 100,\ \sigma=20$' )

plt.axis([20,180,0,0.025])

#añade grid
plt.grid(True)


plt.show
```

```python
# flecha con texto, y fig con otros detalles, el size esta en pulgadas
plt.figure(1,figsize=(10,6),dpi=90)
#solo un grafico
plt.subplot(111)
x = np.arange(0,2*np.pi,0.01)
y=np.cos(x)
plt.plot(x,y, lw=2)

#apuntemos a el valor minimo en pi, putna de la flecha se especifica en xy, el texto en xytext
#shrink es largo flecha

plt.annotate('Mínimo local', 
             xy = (np.pi,-1),
             xytext = (np.pi,-0.5),
             arrowprops = dict(facecolor = "black", shrink = 1.5 ))

plt.ylim(-2.2)
```
